### ●	Detect and treat outliers if necessary
IDENTIFYING OUTLIERS USING IQR (INTER QUANTILE RANGE) METHOD

```python
columns=data_cleaned.select_dtypes(include=['int64','float64']).columns
columns
```

```python
for cols in columns:
    Q1=data_cleaned[cols].quantile(0.25)
    Q3=data_cleaned[cols].quantile(0.75)
    IQR=Q3-Q1
    
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    
    outliers=data[ (data_cleaned[cols]<lower_bound) | (data_cleaned[cols]>upper_bound) ]
    print(f'OUTLIERS IN THE COLUMN {cols}')
    print(outliers[[cols]],'\n')
    
    data_cleaned1=data_cleaned[~(data_cleaned[cols]<lower_bound) | (data_cleaned[cols]>upper_bound)]
```

```python
data_cleaned1.shape
```

```python
data_cleaned2=data_cleaned1.drop(columns=['level_0','index'],errors='ignore')
data_cleaned2
```

```python
data_cleaned2.shape
```
